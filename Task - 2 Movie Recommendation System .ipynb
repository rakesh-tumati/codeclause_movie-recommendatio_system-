```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load the dataset
movies_df = pd.read_csv('movies.csv')

ratings_df = pd.read_csv('ratings.csv')

print(movies_df)
```

```python
print(ratings_df)
```

```python
# Merge the datasets
merged_df = pd.merge(ratings_df, movies_df, on='movieId')
print(merged_df)
```

```python
# Pivot the data to create a user-item matrix
user_movie_matrix = merged_df.pivot_table(index='userId', columns='title', values='rating')
user_movie_matrix
```

```python
# Fill missing values with 0
user_movie_matrix.fillna(0, inplace=True)

# Calculate cosine similarity matrix
cosine_sim_matrix = cosine_similarity(user_movie_matrix)
```

```python
user_movie_matrix
```

```python
def get_movie_recommendations(movie_name, cosine_sim_matrix=cosine_sim_matrix, user_movie_matrix=user_movie_matrix, n=10):
    # Get index of the movie
    movie_index = user_movie_matrix.columns.get_loc(movie_name)
    
    # Get similarity scores for the movie
    movie_similarities = cosine_sim_matrix[movie_index]
    
    # Sort the similarity scores
    movie_similarities_sorted = np.argsort(-movie_similarities)
    
    # Get the top n similar movies
    movie_recommendations = []
    for i in range(n):
        movie_recommendations.append(user_movie_matrix.columns[movie_similarities_sorted[i]])
        
    return movie_recommendations
```
